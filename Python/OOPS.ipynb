Class & Objects:

```python
class Car:
  def __init__(self,window,door,enginetype):
    self.windows=window
    self.doors=door
    self.enginetype=enginetype

  def self_driving(self):
    return "This is a {} car".format(self.enginetype)

obj1 = Car(4,5,"diesel")
print(obj1.windows)
print(obj1.doors)
print(obj1.enginetype)
print(obj1.self_driving())

```

Inheritance -> inherit the objects and access two phases -> parent & child class.

It allows to define a class that inherits all the methods and properties form another class

*   Single Ingeritance
*   Multi-level Inheritance
*  Multiple Inheritance
*  Hirarichal
*   Hybrid







```python
#single Inheritance


class A:
  def method1(self):
    print("A class method is called")

class B(A):
  def method1(self):
    print("B class method is called")

obj1 = B()
obj1.method1()
```

```python
#multi-level inheritnace

class Grandfather:
  def Helping(self):
    print("good in helpingd")

class Father(Grandfather):
  def Caring(self):
    print("good at caring ")

class Son(Father):
  def eat(self):
    print("eating bread.....")

obj1 = Son()
obj1.Caring()
obj1.eat()
obj1.Helping()
```

```python
#multipleinheritance

class Mother:
  mothername = ""
  def mother(self):
    print(self.mothername)

class Father:
  fathername = ""
  def father(self):
    print(self.fathername)

class Son(Mother,Father):
  def parents(self):
    print("Father :",self.fathername)
    print("Mother :",self.mothername)

s1 = Son()
s1.fathername = "Ram"
s1.mothername = "Sita"
s1.parents()
```

```python
#hirarichalinheritance

class Parent:
  def func1(self):
    print("This function is in parent class.")

class Child1(Parent):
  def func2(self):
    print("This function is in child 1.")

class Child2(Parent):
  def func3(self):
    print("This function is in child 2.")

object1 = Child1()
object2 = Child2()
object1.func1()
object1.func2()
object2.func1()
object2.func3()

```

```python
#hybridinheritance

class School:
  def func1(self):
    print("This function is in school.")

class Student1(School):
  def func2(self):
    print("This function is in student 1. ")

class Student2(School):
  def func3(self):
    print("This function is in student 2.")

class Student3(Student1,School):
  def func4(self):
    print("This function is in student 3.")

object = Student3()
object.func1()
object.func2()
```

Polymorphism -> many forms. It refers to use of single method to represent different types in defferent scenarios.


```python
class India():
  def capital(self):
    print("New Delhi is the capital of India.")

  def language(self):
    print("Hindi is the most widely spoken language of India.")

def type(self):
    print("India is a developing country.")

class USA():
  def capital(self):
    print("Washington, D.C. is the capital of USA.")

  def language(self):
    print("English is the primary language of USA.")

def type(self):
    print("USA is a developed country.")

obj_ind = India()
obj_usa = USA()

for country in (obj_ind,obj_usa):
  country.capital()
  country.language()
```

Encapsulation -> means hiding -> wrapping the data and methods that work on that data within one unit

```python
class Base():
  def __init__(self):
    self.a = "Public"
    self._b = "Protected"
    self.__c = "Private"

  def getter(self):
    print(self.__c)

  def setter(self,c):
    self.__c = c
    print(self.__c)

obj1 = Base()
print(obj1.a)
print(obj1._b)
obj1.getter()
obj1.setter("private")
```

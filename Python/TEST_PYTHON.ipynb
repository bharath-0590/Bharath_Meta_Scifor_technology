B3. In Python, exceptions are handled using try, except, else, and finally blocks, which allow you to manage errors gracefully.

```python
''' try: The block of code to be executed, where exceptions might occur.
    except: The block that handles the exception if one is raised.
    else: Executes if no exceptions are raised in the try block.
    finally: Executes code no matter what, whether an exception occurs or not.
    Custom Exception: You can define your own exceptions by subclassing the Exception class.
'''

try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input, please enter a number.")
else:
    print("Division successful:", result)
finally:
    print("Execution completed.")
```

**B6**

Python handles multithreading using the threading module, but due to the Global Interpreter Lock, only one thread can execute Python bytecode at a time, limiting the performance benefits of multithreading for CPU-bound tasks.

Key Differences ->
Multithreading: Uses multiple threads within a single process, sharing memory space, which is effective for I/O-bound tasks.

Multiprocessing: Creates multiple processes, each with its own Python interpreter and memory space, making it suitable for CPU-bound tasks.

```python
#multithreading

import threading
import time

def print_numbers():
    for i in range(5):
        time.sleep(1)
        print(i)

thread = threading.Thread(target=print_numbers)
thread.start()
thread.join()
print("Thread finished execution.")

# multiprocessing

import multiprocessing
import time

def print_numbers():
    for i in range(5):
        time.sleep(1)
        print(i)

if __name__ == "__main__":
    process = multiprocessing.Process(target=print_numbers)
    process.start()
    process.join()
    print("Process finished execution.")

```

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to represent data and methods, promoting modularity and code reusability.

Basic Principles of OOP:

**Encapsulation:** Bundling data and methods within a class, restricting direct access to some of the object's components.

**Abstraction:** Hiding complex implementation details and exposing only the necessary features of an object.

**Inheritance:** bold text Allowing a new class to inherit properties and behaviors from an existing class, promoting code reuse.

**Polymorphism:** Enabling objects to be treated as instances of their parent class, with the ability to override methods for specific behaviors.

Instance variables are specific to each instance of a class, defined within methods using self, while class variables are shared across all instances of a class and are defined directly within the class.

**B8:**

Method overriding occurs when a subclass provides a specific implementation of a method that is already defined in its superclass, changing its functionality, while method overloading involves creating multiple methods with the same name but different parameters within the same class, allowing for different functionalities based on the input.

```python
#method overriding

class Animal:
    def sound(self):
        return "Animal makes a sound"

class Dog(Animal):
    def sound(self):
        return "Dog barks"

my_dog = Dog()
print(my_dog.sound())

#method overloading
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

# Creating an object
calc = Calculator()
print(calc.add(2, 3))
print(calc.add(2, 3, 4))
```

**B9:**

Multiple inheritance is a feature in Python that allows a class to inherit from more than one parent class, enabling the combination of behaviors and attributes from multiple sources.

Python uses the Method Resolution Order (MRO) to determine the order in which classes are looked up when searching for a method. The MRO helps to resolve conflicts when a method is present in multiple parent classes.


```python
class Parent1:
    def greet(self):
        return "Hello from Parent1"

class Parent2:
    def greet(self):
        return "Hello from Parent2"

class Child(Parent1, Parent2):
    pass

child = Child()
print(child.greet())

#MRO
print(Child.__mro__)
```

**B10:**  

Abstract classes are classes that cannot be instantiated and are meant to be subclassed, providing a common interface for derived classes. They can define abstract methods that must be implemented by any non-abstract subclass.

Interfaces can be seen as a type of abstract class that only defines methods without any implementation.

The abc module provides the ABC class and the abstractmethod decorator to define abstract base classes.

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def area(self):  # Abstract method
        pass

class Circle(Shape):  # Concrete class
    def __init__(self, radius):
        self.radius = radius

    def area(self):  # Implementing abstract method
        return 3.14 * self.radius ** 2

# Creating an object of Circle
circle = Circle(5)
print(circle.area())
```

**B12**

Encapsulation in OOP is the concept of restricting direct access to an object's internal state and requiring all interaction to be performed through well-defined methods.

Encapsulation hides internal details by making certain attributes private and only allowing access through public methods, ensuring controlled interaction with an object's state.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0  # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount  # modify private variable

    def get_balance(self):  # public method to access private variable
        return self.__balance

account = BankAccount()
account.deposit(100)

print(account.get_balance())
```

**B14**

Operator overloading in Python allows you to define custom behavior for operators -> +, -, etc., for user-defined classes by implementing special methods such as __add__, __sub__, etc.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(5, 7)

# Using overloaded operators
p3 = p1 + p2
p4 = p1 - p2

print(p3)
print(p4)
```

```python
'''Q1:data for handling missing values and then drawing bar and pie charts'''


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset from the specified path
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Meta Scifor Technology\StudentPerformanceFactors.csv")

# Display the first few rows
pd.set_option('display.max_columns',50)
df
```

```python
df.info()
```

```python
df = df.replace('?',np.NaN)
```

```python
df
```

```python
df.info()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])
```

```python
# Remove rows with missing values
df_cleaned = df.dropna()
```

```python
# Check again for any missing values
print("Missing values after cleaning:")
print(df_cleaned.isnull().sum())
```

```python
# Bar chart for 'StudyTime' column
plt.figure(figsize=(9, 6))
```

```python
sns.countplot(data=df_cleaned, x='Hours_Studied')
plt.title('Bar Chart of Hours_Studied')
plt.xlabel('Hours_Studied')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for 'StudyTime' column
study_time_counts = df_cleaned['Hours_Studied'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(study_time_counts, labels=study_time_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Hours_Studied')
plt.show()
```

```python
# Bar chart for 'Attendance' column 

plt.figure(figsize=(8, 6))
sns.countplot(data=df_cleaned, x='Attendance')
plt.title('Bar Chart of Attendance')
plt.xlabel('Attendance')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart for 'attendance' column
attendance_counts = df_cleaned['Attendance'].value_counts()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(attendance_counts, labels=attendance_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Attendance')
plt.show()
```

 Q2: Use of pandas, numpy, seaborn, matplotlib? Explain with example

 Pandas-> commonly used for data cleaning, transformation and analysis

```python
import pandas as pd
```

```python
# Creating a DataFrame

data = {
    'Name': ['Kumar', 'Vijay', 'John'],
    'Age': [24, 30, 22],
    'City': ['Telangana', 'Andhra Pradesh', 'Maharastra']
}
df = pd.DataFrame(data)
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
df.info()
```

```python
# Data Cleaning - Checking for missing values
df.isnull().sum()
```

```python
#Transformation - Adding a new column 'Age in 5 years'
df['Age in 5 years'] = df['Age'] + 5
```

```python
df
```

```python
#Data Analysis - Filtering people with age greater than 25
filtered_df = df[df['Age'] > 25]
```

```python
filtered_df
```

Numpy -> It is widely used for scientific computing and performing operations on arrays.

```python
import numpy as np

# Creating a NumPy array
array = np.array([1, 2, 3, 4, 5])

# Performing operations
squared = array ** 2
print(squared)
```

Seaborn -> It is used for visualizing complex datasets with ease, including the ability to create informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset('tips')

# Creating a scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.title('Total Bill vs Tip')
plt.show()
```

Matplotlib - > It is commonly used for creating plots, histograms, bar charts, and other visual representations of data.

```python
import matplotlib.pyplot as plt

# Creating a simple line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
